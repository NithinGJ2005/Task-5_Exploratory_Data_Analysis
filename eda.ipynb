```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

#Load dataset
df = pd.read_csv("/content/titanic.csv")
df.head()

#Basic Information
df.info()
df.describe()

df['Survived'].value_counts()
df['Sex'].value_counts()
df['Pclass'].value_counts()

```

```python
#Visualizations
#Survival Count
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

Observation: More passengers died than survived.

```python
#Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

Observation: Females had significantly higher survival rate.

```python
#Survival by Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

Observation: 1st class passengers survived more.

```python
#Age Distribution
plt.hist(df['Age'].dropna(), bins=20)
plt.title("Age Distribution")
plt.show()
```

Observation: Majority passengers were between 2040 years.

```python
#Boxplot (Age vs Survival)
sns.boxplot(x='Survived', y='Age', data=df)
plt.show()
```

Observation: Younger passengers had slightly better survival rate.

```python
#Pairplot
sns.pairplot(df[['Survived','Age','Fare','Pclass']], hue='Survived')
plt.show()
```

Observation: Fare and Pclass strongly influence survival.

```python
#Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

Observation:

Fare positively correlated with survival

Pclass negatively correlated with survival
